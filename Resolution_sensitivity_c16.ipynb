```python
import matplotlib.pyplot as plt # plotting
import numpy as np              # working with number arrays etc.
from netCDF4 import Dataset     # reading netcdf files

pathles="/Users/amrapallig/Desktop/cvmix/"
path="/Users/amrapallig/Desktop/cvmix/adc/convection/"

path2="/Users/amrapallig/run/cvmix_test/forward/output/convection/time_sensitivity/"
```

```python
fig = plt.figure(figsize=(13,11))
t=96
t1=t-6
mks=15
mks1=12
lw=3

d=70
plt.subplot(221)

lesc1= Dataset(pathles+"les/Cooling16.nc", mode='r')

fh_1m=Dataset(path+'output.0001-01-01_00.00_c2_1m_test2.nc', mode='r')  #_6 is repo 6, _4 is repo 5 with w3check-0 from 1
fh_2m=Dataset(path+'output.0001-01-01_00.00_c2_2m_test2.nc', mode='r')   #_3 is with 1,1.5, _2 is with 10,15
fh_5m=Dataset(path+'output.0001-01-01_00.00_c2_5m_test2.nc', mode='r')
fh_10m=Dataset(path+'output.0001-01-01_00.00_c2_10m_test2.nc', mode='r')

mfc1_1m = Dataset(path+'adcTurbulenceArrays_c16_1m_test3.nc', mode='r')  
mfc1_2m = Dataset(path+'adcTurbulenceArrays_c16_2m_test2.nc', mode='r')  
mfc1_5m = Dataset(path+'adcTurbulenceArrays_c16_5m_test2.nc', mode='r')
mfc1_10m = Dataset(path+'adcTurbulenceArrays_c16_10m_test2.nc', mode='r')




plt.plot(np.mean(lesc1.variables["temp"][t1:t,:],axis=0),lesc1.variables["zm"][:],'-k',linewidth=lw)
plt.plot(np.nanmean(mfc1_1m.variables['temperature'][t1:t,0,:],axis=0),fh_1m.variables['zMid'][1,1,:],'--r',linewidth=lw)
plt.plot(np.nanmean(mfc1_2m.variables['temperature'][t1:t,0,:],axis=0),fh_2m.variables['zMid'][1,1,:],'--b',linewidth=lw)
plt.plot(np.nanmean(mfc1_5m.variables['temperature'][t1:t,0,:],axis=0),fh_5m.variables['zMid'][1,1,:],':*c',markersize=mks)
plt.plot(np.nanmean(mfc1_10m.variables['temperature'][t1:t,0,:],axis=0),fh_10m.variables['zMid'][1,1,:],':pm',markersize=mks1)
#plt.legend(['les','adc_1m','adc_2m','adc_5m','adc_10m'],loc=4)
plt.ylim(-d,1)
plt.xlim(13,20)
plt.grid()
plt.legend(['LES','ADC_1m','ADC_2m','ADC_5m','ADC_10m'],loc=1,fontsize=14)


#plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))

plt.xlabel('$\overline{\\theta}$ $(^0C)$ ',fontsize=20)
plt.ylabel('$Depth$ $(m)$ ',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.text(18.5, -55,"(a)",fontsize=20)


plt.subplot(222)

plt.plot(np.mean(lesc1.variables["wt"][t1:t,:],axis=0),lesc1.variables["zm"][:],'-k',linewidth=lw)
plt.plot(np.nanmean(mfc1_1m.variables['wt'][t1:t,0,:-1]+mfc1_1m.variables['wt_spsU'][t1:t,0,:-1],axis=0),fh_1m.variables['zTop'][1,1,:],'--r',linewidth=lw)
plt.plot(np.nanmean(mfc1_2m.variables['wt'][t1:t,0,:-1]+mfc1_2m.variables['wt_spsU'][t1:t,0,:-1],axis=0),fh_2m.variables['zTop'][1,1,:],':b',linewidth=lw)
plt.plot(np.nanmean(mfc1_5m.variables['wt'][t1:t,0,:-1]+mfc1_5m.variables['wt_spsU'][t1:t,0,:-1],axis=0),fh_5m.variables['zTop'][1,1,:],':*c',markersize=mks)
plt.plot(np.nanmean(mfc1_10m.variables['wt'][t1:t,0,:-1]+mfc1_10m.variables['wt_spsU'][t1:t,0,:-1],axis=0),fh_10m.variables['zTop'][1,1,:],':pm',markersize=mks1)
plt.ylim(-d,1)
plt.xlabel('$\overline{w^\prime\\theta^\prime}$ $(K$ $m s^{-1})$',fontsize=20)
plt.grid()
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.text(1.5e-4, -55,"(b)",fontsize=20)


plt.subplot(223)

plt.plot(np.mean(lesc1.variables["w2"][t1:t,:],axis=0),lesc1.variables["zm"][:],'-k',linewidth=lw)
plt.plot(np.nanmean(mfc1_1m.variables['w2'][t1:t,0,:-1],axis=0),fh_1m.variables['zTop'][1,1,:],'--r',linewidth=lw)
plt.plot(np.nanmean(mfc1_2m.variables['w2'][t1:t,0,:-1],axis=0),fh_2m.variables['zTop'][1,1,:],':b',linewidth=lw)
plt.plot(np.nanmean(mfc1_5m.variables['w2'][t1:t,0,:-1],axis=0),fh_5m.variables['zTop'][1,1,:],':*c',markersize=mks)
plt.plot(np.nanmean(mfc1_10m.variables['w2'][t1:t,0,:-1],axis=0),fh_10m.variables['zTop'][1,1,:],':pm',markersize=mks1)
#plt.legend(['les','adc_1m','adc_2m','adc_5m','adc_10m'],loc=4)
plt.ylim(-d,1)
plt.xlabel('$\overline{{w^\prime}^2}$  $(m^2/s^2)$ ',fontsize=20)
plt.grid()
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('$Depth$ $(m)$ ',fontsize=20)
plt.text(2.5e-4, -55,"(c)",fontsize=20)


plt.subplot(224)

plt.plot(np.mean(lesc1.variables["w3"][t1:t,:],axis=0),lesc1.variables["zm"][:],'-k',linewidth=lw)
plt.plot(np.nanmean(mfc1_1m.variables['w3'][t1:t,0,:-1],axis=0),fh_1m.variables['zMid'][1,1,:-1],'--r',linewidth=lw)
plt.plot(np.nanmean(mfc1_2m.variables['w3'][t1:t,0,:-1],axis=0),fh_2m.variables['zMid'][1,1,:-1],':b',linewidth=lw)
plt.plot(np.nanmean(mfc1_5m.variables['w3'][t1:t,0,:-1],axis=0),fh_5m.variables['zMid'][1,1,:-1],':*c',markersize=mks)
plt.plot(np.nanmean(mfc1_10m.variables['w3'][t1:t,0,:-1],axis=0),fh_10m.variables['zMid'][1,1,:-1],':pm',markersize=mks1)
#plt.legend(['les','adc_1m','adc_2m','adc_5m','adc_10m'],loc=4)
plt.ylim(-d,1)
plt.xlabel('$\overline{{w^\prime}^3}$ $(m^3/s^3)$ ',fontsize=20)
plt.grid()
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.text(3e-6, -55,"(d)",fontsize=20)
```
